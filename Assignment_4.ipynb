```python
import pandas as pd
import numpy as np
import re
```

```python
df2018 = pd.read_csv('FIFA-2018.csv')
df2022 = pd.read_csv('FIFA-2022.csv')
```

```python
df2022.info()
```

```python
# Dropping Columns to make an easier comparison
df2018.drop(['Goals For', 'Goals Against', 'Goal Difference'], axis=1, inplace=True)
df2022.drop(['Goals For', 'Goals Against', 'Goal Difference'], axis=1, inplace=True)
```

```python
# Change The Data Name
for columnName in df2018.iteritems():
    newColumnName = "".join(columnName[0]) + " 2018"
    df2018.rename(columns = {columnName[0] : newColumnName}, inplace = True)
for columnName in df2022.iteritems():
    newColumnName = "".join(columnName[0]) + " 2022"
    df2022.rename(columns = {columnName[0] : newColumnName}, inplace = True)
    
# Change Index
df2018.set_index('Team 2018', inplace=True)
df2022.set_index('Team 2022', inplace=True)
```

```python
# Melt
df2018_view = pd.melt(df2018, id_vars=['Position 2018','Games Played 2018'], value_vars=['Points 2018'], var_name='Attribute', value_name='Value')
df2018_view
```

```python
# Pivot
df2018_pivot = df2018_view.pivot(index=['Position 2018','Games Played 2018'], columns='Attribute')
df2018_pivot = df2018_pivot['Value'].reset_index()
df2018_pivot.columns.name = None
df2018_pivot
```

```python
# Merge
df1822 = df2018.merge(df2022, left_index=True, right_index=True, how='outer')
df1822
```

```python
import requests
from io import StringIO
url = 'https://raw.githubusercontent.com/nnqomariyah/Fundamentals_of_Data_Science/main/week_3/data/movies_metadata.csv'
response = requests.get(url)
data = StringIO(response.text)
df_movie_unclean = pd.read_csv(data)

df_movie_unclean
```

```python
df_movie_unclean
```

```python
# Clean Collection
for collection in range(len(df_movie_unclean['belongs_to_collection'])):
    text = df_movie_unclean._get_value(collection,'belongs_to_collection')
    saved = text
    if type(text) == str:
        text = re.sub('id|name|poster_path|backdrop_path|\'|\"|\:|,|\"|\}|\{|[0-9]', '', text)
        true_text = ""
        for t in text.split():
            r = t
            r = re.sub('^/.*jpg$', '', r)
            true_text = true_text + " " + r
        df_movie_unclean['belongs_to_collection'] = df_movie_unclean['belongs_to_collection'].replace([saved], true_text)
```

```python
# Clean Genres
for genre in range(len(df_movie_unclean['genres'])):
    text2 = df_movie_unclean._get_value(genre,'genres')
    saved2 = text2
    text2 = re.sub('id|name|\'|\"|\:|,|\"|\}|\{|\]|\[|[0-9]', '', text2)
    true_text2 = ""
    for t2 in text2.split():
        true_text2 = true_text2 + ", " + t2
    df_movie_unclean['genres'] = df_movie_unclean['genres'].replace([saved2], true_text2)
```

```python
# Clean Spoken Languange
for lang in range(len(df_movie_unclean['spoken_languages'])):
    text3 = df_movie_unclean._get_value(lang,'spoken_languages')
    saved3 = text3
    if type(text3) == str:
        text3 = re.sub('iso_639_1|name|\'|\"|\:|,|\"|\}|\{|\]|\[|[0-9]', '', text3)
        true_text3 = ""
        for t3 in text3.split():
            r3 = t3
            r3 = re.sub('^..$', '', r3)
            true_text3 = true_text3 + ", " + r3
        df_movie_unclean['spoken_languages'] = df_movie_unclean['spoken_languages'].replace([saved3], true_text3)
```
